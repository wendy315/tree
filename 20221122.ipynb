```python
def read_dataset(fname):
     
     # 指定索引
    data = pd.read_csv(fname, index_col=False) 
    labels = data['classification'].unique().tolist()
    data['classification'] = data['classification'].apply(lambda n: labels.index(n))
    data['rbc'] = [i for i in data['rbc'].replace(['normal','abnormal'],[1,2])] #1:normal 2:abnormal
    data['pc'] = [i for i in data['pc'].replace(['normal','abnormal'],[1,2])] #1:normal 2:abnormal
    data['pcc'] = [i for i in data['pcc'].replace(['notpresent','present'],[1,2])] #1:notpresent 2:present
    data['ba'] = [i for i in data['ba'].replace(['notpresent','present'],[1,2])] #1:notpresent 2:present
    data['bu'] = [float(i) for i in data['bu'].replace('?',0)]
    data['sc'] = [float(i) for i in data['sc'].replace('?',0)]
    
    data['htn'] = [i for i in data['htn'].replace(['yes','no'],[1,2])] #1:yes 2:no
    data['dm'] = [i for i in data['dm'].replace(['yes','no'],[1,2])] #1:yes 2:no
    data['cad'] = [i for i in data['cad'].replace(['yes','no'],[1,2])] #1:yes 2:no
    data['appet'] = [i for i in data['appet'].replace(['good','poor'],[1,2])] #1:good 2:poor
    data['pe'] = [i for i in data['pe'].replace(['yes','no'],[1,2])] #1:yes 2:no
    data['ane'] = [i for i in data['ane'].replace(['yes','no'],[1,2])] #1:yes 2:no
    
    # 處理遺缺值
    #data = data.fillna(0)
    return data
data = read_dataset('kidney_disease3.csv')
```

```python
#有些欄位應該是數值，卻是object，因為有空值(nan)造成，以下三個欄位是肉眼觀察到的
mistyped=['rbc','pcv','wc','rc']
numeric=[]

for i in data.columns:
    if data[i].dtype=='float64':
        numeric.append(i)

#將float64的欄位與肉眼觀察到的欄位合起來，都應該是numberic才對
numeric=numeric+mistyped
```

```python
#看看數值型欄位
numeric
```

```python
#轉換為float
for col in mistyped:
        data[col]=data[col].astype('float')
```

```python
#接下來處理類別欄位，且排除掉我們的數值型欄位
categoricals=['dm','cad']

for col in data.columns:
    if not col in numeric:
        categoricals.append(col)

#排除掉最後一個應變數
categoricals.remove('classification')
```

```python
#看看類別欄位
categoricals
```

```python
#類別型欄位塞眾數
for col in categoricals:
    data[col]=data[col].fillna(data[col].mode()[0])
#數值型欄位塞中位數
for col in numeric:
    data[col]=data[col].fillna(data[col].mean())
#再次檢查欄位遺缺值狀態
data.isnull().any()
```

```python
from sklearn.model_selection import train_test_split

y = data['classification'].values
X =  data.drop(['classification'], axis=1).values

#切分訓練與驗證資料集
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 100, test_size = 0.2)

print('train dataset: {0}; test dataset: {1}'.format(
    X_train.shape, X_test.shape))
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',max_depth = 3) 
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
#模型準確度
y_pred = model.predict(X_test)
print("Train data accuracy:",accuracy_score(y_true = y_train, y_pred=model.predict(X_train)))
print("Test data accuracy:",accuracy_score(y_true = y_test, y_pred=y_pred))
#測試不可能比訓練來得好
```
